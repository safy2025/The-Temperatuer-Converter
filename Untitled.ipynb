```python
while 1>0:
    print("Welcome to the Temperatuer Converter!")
    Type = input(" Do you want to convert from  Celsius to Fahrenheit (C) or Fahrenheit  to Celsius (F) ? ")
    
    if Type == "Celsius to Fahrenheit (C)":
        temp= float(input("Enter The Temperature in  Celsius : "))
        F= (temp *(9/5) )+32
        print ("The Temperature in Fahrenheit",F,"F ")
    elif Type == "Fahrenheit  to Celsius (F)" :
        temp= float(input("Enter The Temperature in Fahrenheit : "))
        C= (temp -32) *(5/9)
        print ("The Temperature in  Celsius",C,"C")
    repeat=input("Do you like to perform another conversion? (yes/no):")
    if repeat !="yes":
        print("Thank you for using the Temperatuer Converter!")
        break
```
